```python
pdf_prod = pdf_gearvn[["Ngành", "Hãng", "Mã sản phẩm", "Kho lấy hàng","Chiết khấu", 
 "Giá", "Tiền chiết khấu"]].drop_duplicates()
print(f"Number of products: {pdf_prod.shape[0]}")
display(pdf_prod.head())

print_runtime()
```

```python
pdf_prod["Chiết khấu"].fillna(0, inplace=True)
display(pdf_prod.head())
```

```python
#Replace "," to ""
pdf_prod["Chiết khấu"].replace(to_replace =",",
                 value ="", inplace = True)
display(pdf_prod.head())
```

```python
pdf_prod["Chiết khấu"] = pdf_prod["Chiết khấu"].astype(float)
```

```python
pdf_prod["Tiền chiết khấu"].fillna(0, inplace=True)
display(pdf_prod.head())
```

```python
#Replace "-" to ""
pdf_prod["Giá"].replace(to_replace ="-",
                 value ="0", inplace = True)
display(pdf_prod.head())
```
