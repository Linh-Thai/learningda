```python
x = str(input('Fill Product Name: '))
print(x)
```

```python
pdf_gg = pdf_orderdetail[(pdf_orderdetail["Mã sản phẩm"] == x)]
display(pdf_gg.head())
print(pdf_gg.shape)
```

```python
#Numbers of product sold
numbers_gg_sold = pdf_gg.shape[0]
```

```python
pdf_lap = pdf_orderdetail[(pdf_orderdetail["Ngành"] == 'LAP')]
display(pdf_lap.head())
print(pdf_lap.shape)
```

```python
ls_col_lap_gg = ['ID tích hợp', 'Mã sản phẩm_gg', 'Mã sản phẩm_lap', 'Số lượng_gg', 'Số lượng_lap']
pdf_lap_with_gg = (pdf_gg.merge(pdf_lap, on = 'ID tích hợp', how = 'left', suffixes = ('_gg','_lap')))[ls_col_lap_gg]
display(pdf_lap_with_gg.head())
print(pdf_lap_with_gg.shape)
```

```python
ls_col_pc = ['ID tích hợp','Số lượng_mb']
pdf_cpu = pdf_orderdetail[(pdf_orderdetail["Ngành"] == 'CPU')]
pdf_mb = pdf_orderdetail[(pdf_orderdetail["Ngành"] == 'MB')]
pdf_psu = pdf_orderdetail[(pdf_orderdetail["Ngành"] == 'PW')]
pdf_pc = (pdf_cpu.merge(pdf_mb, on = 'ID tích hợp', how = 'inner', suffixes=('_cpu','_mb'))
          .merge(pdf_psu,on = 'ID tích hợp', how = 'inner'))\
[ls_col_pc]
pdf_pc.columns = ['ID tích hợp','Số lượng_pc']
display(pdf_pc.head())
print(pdf_pc.shape)
```

```python
ls_col_pc_gg = ['ID tích hợp', 'Mã sản phẩm_gg', 'Mã sản phẩm_pc', 'Ngành_gg', 'Ngành_pc', 'Số lượng_gg', 'Số lượng_pc']
pdf_pc_lap_with_gg = pdf_lap_with_gg.merge(pdf_pc, on = 'ID tích hợp', how = 'left', suffixes = ('_gg','_lap'))

display(pdf_pc_lap_with_gg.head())
print(pdf_pc_lap_with_gg.shape)
```

```python
lap_with_gg_ratio = (pdf_pc_lap_with_gg['Số lượng_lap'].sum()) / numbers_gg_sold *100
pc_with_gg_ratio = (pdf_pc_lap_with_gg['Số lượng_pc'].sum()) / numbers_gg_sold *100
only_gg_ratio = 100 - lap_with_gg_ratio - pc_with_gg_ratio
print('Gaming Gear with Laptop Ratio: {:.2f}%'.format(lap_with_gg_ratio))
print('Gaming Gear with PC Ratio: {:.2f}%'.format(pc_with_gg_ratio))
print('Only Gaming Gear Ratio: {:.2f}%'.format(only_gg_ratio))
```
